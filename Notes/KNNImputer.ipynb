### 🧠 KNNImputer — Summary

#### 💡 What It Does
`KNNImputer` fills missing values by finding **K most similar rows (neighbors)** and using their **average value** for the missing cell.

---

#### ⚙️ How It Finds “Similar” Rows
- “Similar” means **rows that look alike** in terms of other column values — not rows that are above or below in the dataset.  
- It measures similarity based on **distance** (usually Euclidean distance).  
- Typically, `k = 5` → uses 5 most similar rows.

---

#### 🧩 When a Column Has a Missing Value
Example:  
If column **A** has a missing value in a row:
1. The imputer looks at all **other known columns** in that row (say B, C, D, …).  
2. Finds **K nearest rows** where A and those columns are not missing.  
3. Takes the **average** of those K rows’ A values.  
4. Fills that average into the missing spot.

If **B** has a missing value, the same process repeats — but using A, C, D, etc. to find similar rows.

---

#### 📘 Simple Example

Suppose we have 2 columns: **A** and **B**

| Row | A | B  |
|------|---|----|
| 1 | 1 | 10 |
| 2 | 2 | 12 |
| 3 | 3 | NaN |
| 4 | 4 | 18 |
| 5 | NaN | 20 |
| 6 | 6 | 25 |
| 7 | 7 | NaN |
| 8 | 8 | 30 |
| 9 | 9 | 35 |
| 10 | 10 | 40 |

Now:
- For **Row 5 (A = NaN)** → The imputer uses **B = 20** to find 5 most similar rows (based on B).  
  Then it averages those rows’ **A** values and fills it in.
  
- For **Row 7 (B = NaN)** → The imputer uses **A = 7** to find 5 most similar rows (based on A).  
  Then it averages those rows’ **B** values and fills it in.

---

#### 📊 Key Points to Remember
- “Neighbors” = **most similar rows**, not physically near ones.  
- Each missing cell is handled **individually**.  
- Uses **mean of neighbors’ values** (by default).  
- Works best when features have similar scales (if not, scaling is usually done first).

---

#### ✅ In One Line
> `KNNImputer` fills a missing value by finding K most similar rows based on other known columns in that row and replaces the missing value with the average of those neighbors’ values.
